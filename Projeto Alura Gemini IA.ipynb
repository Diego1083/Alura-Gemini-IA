```python

from geopy.geocoders import Nominatim
from geopy.distance import geodesic

def calcular_distancia_e_gasto():
    """
    Calcula a distância entre duas cidades e estima o gasto com gasolina para uma viagem de ida e volta,
    usando a biblioteca geopy para calcular a distância.
    """

    print("Bem-vindo ao Calculador de Distância e Gasto com Combustível!")
    print("Este programa calcula a distância entre duas cidades e estima o gasto com gasolina para uma viagem de ida e volta.")
    print("Por favor, insira as cidades de origem e destino.")
    # Solicita ao usuário as cidades de origem e destino
    cidade_origem = input("Digite o nome da cidade de origem: ")
    cidade_destino = input("Digite o nome da cidade de destino: ")

    # Inicializa o geolocator
    geolocator = Nominatim(user_agent="calculo_distancia")

    try:
        # Obtém as coordenadas das cidades
        location_origem = geolocator.geocode(cidade_origem)
        location_destino = geolocator.geocode(cidade_destino)

        if location_origem is None or location_destino is None:
            print("Erro: Não foi possível encontrar as coordenadas de uma ou ambas as cidades.")
            return

        # Extrai as coordenadas (latitude e longitude)
        coord_origem = (location_origem.latitude, location_origem.longitude)
        coord_destino = (location_destino.latitude, location_destino.longitude)

        # Calcula a distância usando geodesic
        distancia_ida = geodesic(coord_origem, coord_destino).km
        distancia_volta = distancia_ida
        distancia_total = distancia_ida + distancia_volta

        print(f"A distância de {cidade_origem} para {cidade_destino} é de {distancia_total:.2f} km (ida e volta).")

        # Solicita ao usuário o consumo de combustível do veículo
        consumo_combustivel_str = input("Digite o consumo de combustível do seu veículo (km/litro): ")
        # Substitui a vírgula por ponto para a conversão correta para float
        consumo_combustivel = float(consumo_combustivel_str.replace(',', '.'))

        # Solicita ao usuário o preço da gasolina
        preco_gasolina_str = input("Digite o preço do litro da gasolina (R$): ")
        # Substitui a vírgula por ponto para a conversão correta para float
        preco_gasolina = float(preco_gasolina_str.replace(',', '.'))

        # Calcula o gasto com gasolina
        litros_necessarios = distancia_total / consumo_combustivel
        gasto_total = litros_necessarios * preco_gasolina

        print(f"O gasto estimado com gasolina para a viagem é de R$ {gasto_total:.2f}")
        print("Obrigado por usar o Calculador de Distância e Gasto com Combustível!")


    except Exception as e:
        print(f"Ocorreu um erro: {e}")

if __name__ == "__main__":
    calcular_distancia_e_gasto()
```
